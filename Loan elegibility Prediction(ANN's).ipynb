# Define X and Y

```python
Y=A[["Loan_Status"]]
X=A.drop(labels=["Loan_Status","Loan_ID"],axis=1)
```

# Encoding the Y

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y.Loan_Status=le.fit_transform(Y)            #
```

```python
from preprocessing_6pm import data_prep
xnew=data_prep(X)
```

# Divide the Data into Training and Testing

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(xnew,Y,test_size=0.2,random_state=21)
```

# Create NN

```python
from keras.layers import Dense,Dropout
from keras.models import Sequential
from keras.callbacks import EarlyStopping


nn=Sequential()
es=EarlyStopping(patience=50)
nn.add(Dense(200,input_dim=(xnew.shape[1])))
nn.add(Dense(200))
nn.add(Dropout(0.1))
nn.add(Dense(200))
nn.add(Dropout(0.1))
nn.add(Dense(1,activation="sigmoid"))
```

```python
nn.compile(loss="binary_crossentropy",metrics=["accuracy"])
model=nn.fit(xtrain,ytrain,epochs=100,validation_split=0.2,batch_size=4,callbacks=es)
```
